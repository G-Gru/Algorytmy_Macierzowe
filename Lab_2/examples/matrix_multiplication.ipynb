A file for example usages of matrix multiplication algorithm.

```julia
include(joinpath(@__DIR__, "..", "lib", "strassen_multiplication.jl"))
include(joinpath(@__DIR__, "..", "lib", "strassen_padded.jl"))
using Test
using .strassen_multiplication
using .strassen_padded
```

```julia
A = rand(4, 4)
B = rand(4, 4)
C_ref = A * B
println("Testing Strassen multiplication without padding...")
C_strassen, add_count, mul_count = strassen(A, B)
@test isapprox(C_strassen, C_ref; atol=1e-10)
println("Addition operations: $add_count, Multiplication operations: $mul_count")
println("expected result:")
println(C_ref)
println("Strassen result:")
println(C_strassen)
```

```julia
A = rand(5, 5)
B = rand(5, 5)
C_ref = A * B
println("\nTesting Strassen multiplication with padding...")
C, add_count_pad, mul_count_pad = strassen_pad(A, B)
@test isapprox(C, C_ref; atol=1e-10)
println("Addition operations: $add_count_pad, Multiplication operations: $mul_count_pad")
println("expected result:")
println(C_ref)
println("Strassen padded result:")
println(C)
```

```julia
A = rand(4, 6)
B = rand(6, 7)
C_ref = A * B
println("\nTesting Strassen multiplication with padding for non square matrices...")
C, add_count_pad, mul_count_pad = strassen_pad(A, B)
@test isapprox(C, C_ref; atol=1e-10)
println("Addition operations: $add_count_pad, Multiplication operations: $mul_count_pad")
println("expected result:")
println(C_ref)
println("Strassen padded result:")
println(C)
```
