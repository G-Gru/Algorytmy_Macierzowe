```julia
include("../lib/show_image.jl")
include("../lib/show_compression.jl")
include("../src/Lab_3.jl")
include("../src/Lab_4.jl")

using FileIO, Colors, ImageCore, ImageShow, ImageMagick, Serialization

using .Lab_3
using .Lab_4
using .show_image
using .show_compression
using LinearAlgebra
using BenchmarkTools.@belapsed
```

```julia
# parametry
N = 20
k = 3                # mała liczba
max_rank = 4
max_error = 1e-4
compressed_times = Float64[]
normal_times = Float64[]
ns = 1:N

for n in ns
    println("n = $n")

    # generowanie macierzy
    A = gen_network_matrix(n)
    C = compression_tree(A, 1, n, 1, n, max_rank, max_error)

    # rozmiar wektora
    dim = 2^(3n)
    x = rand(dim)

    # czas: mnożenie skompresowane
    t_comp = @belapsed matrix_vector_mult($C, $x)
    push!(compressed_times, t_comp)

    # czas: mnożenie klasyczne
    t_norm = @belapsed $A * $x
    push!(normal_times, t_norm)
end

# wykres
plot(
    ns,
    compressed_times,
    label = "Mnożenie skompresowane",
    xlabel = "n",
    ylabel = "Czas [s]",
    lw = 2,
    marker = :circle
)

plot!(
    ns,
    normal_times,
    label = "Mnożenie klasyczne",
    lw = 2,
    marker = :square
)
```
